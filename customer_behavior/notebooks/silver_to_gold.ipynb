```python
import os
```

```python
def executar_sqls():
  # percorre todos os arquivos .sql da pasta
  for arquivo in os.listdir():
      if arquivo.endswith(".sql"):
        # lê o conteúdo do script
        with open(arquivo, "r", encoding="utf-8") as f:
            sql_script = f.read()

            # executa o script e carrega em DataFrame
            df = pd.read_sql(sql_script, engine)

            # monta nome do arquivo de saída (mesmo nome do .sql, mas .csv)
            nome_saida = os.path.splitext(arquivo)[0] + ".csv"

            # salva em CSV
            df.to_csv(nome_saida, index=False)
            print(f"✅ {arquivo} executado e salvo.")
```

```python
executar_sqls()
```
